```python
#script to get the location and brief of each university using bs4
import pandas as pd
import requests
from bs4 import BeautifulSoup

csv_file = "us_uni_urls.csv"
df = pd.read_csv(csv_file)

#CSS selectors location and description
LOCATION_CSS_CLASS = "css-9zlu29"  
DESCRIPTION_CSS_CLASS = "css-1xxxan3"  

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3",
    "Accept": "application/json",
    "Referer": "https://www.timeshighereducation.com/world-university-rankings/2025/world-ranking",

}

def get_university_info(url):
    try:
        response = requests.get(url,headers=headers, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        location_tag = soup.find('div', class_="css-9zlu29")
        description_tag = soup.find(class_=DESCRIPTION_CSS_CLASS)
        
        location = location_tag.get_text(strip=True) if location_tag else "N/A"
        description = description_tag.get_text(strip=True) if description_tag else "N/A"
        
        return location, description
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return "N/A", "N/A"


data = []

for url in df.iloc[:, 0]: 
    location, description = get_university_info(url)
    data.append([url, location, description])

# Create DataFrame and save results
result_df = pd.DataFrame(data, columns=["URL", "Location", "Description"])
result_df.to_csv("university_info.csv", index=False)

print("Scraping completed. Data saved to university_info.csv.")
```

### Getting the urls and short descriptions

```python
#script to get the university urls using google serp api and retrieveing the first result by searching by name from the us_uni_names.csv dataset
import pandas as pd
import requests
import time
import os
# Constants
  # Replace with your actual SERP API key
API_KEY = os.getenv("SERP_API_KEY")
SERP_API_URL = "https://serpapi.com/search"
CSV_FILE = "us_uni_names.csv"
OUTPUT_FILE = "university_urls.csv"

def get_university_url(query: str) -> str:
    """Fetches the first search result URL for a given university name."""
    params = {
        "engine": "google",
        "q": query,
        "api_key": API_KEY,
        "num": 1  # Get only the first result
    }
    response = requests.get(SERP_API_URL, params=params)
    if response.status_code == 200:
        data = response.json()
        if "organic_results" in data and data["organic_results"]:
            return data["organic_results"][0]["link"]  # Extract first result URL
    return "Not Found"

```

```python
df = pd.read_csv(CSV_FILE)
    
if "name" not in df.columns:
    print("Error: 'name' column not found in CSV file.")

df["uni_url"] = df["name"].apply(get_university_url)
df.to_csv(OUTPUT_FILE, index=False)
print(f"Results saved to {OUTPUT_FILE}")
```

```python
import pandas as pd
from langchain_groq import ChatGroq
from langchain.schema import SystemMessage, HumanMessage

# Set up Groq API
groq_api_key = os.getenv("GROQ_API_KEY")  # Replace with your actual Groq API key
llm = ChatGroq(model="llama-3.3-70b-specdec", temperature=0.7, groq_api_key=groq_api_key)

# Load CSV file
df = pd.read_csv("uni_with_description.csv")  # Ensure your CSV is correctly formatted
counter = 1

def expand_description(row):
    """Uses Groq LLM via LangChain to generate a detailed university description."""
    prompt = f"""
    Provide a detailed description for {row['name']} located in {row['Location']}.
    This university is ranked {row['rank']} globally.

    Key information:
    - Citations Score: {row['scores_citations']}
    - Research Score: {row['scores_research']}
    - Teaching Score: {row['scores_teaching']}
    - Student Population: {row['stats_number_students']}
    - International Students: {row['stats_pc_intl_students']}
    - Subjects Offered: {row['subjects_offered']}

    Write the response in a **formal and informative tone**, focusing on academic excellence, campus life, global reputation, and research impact.
    """
    print("Processing:")
    x  = list(row)
    response = llm.invoke([SystemMessage(content="You are an expert education consultant."), HumanMessage(content=prompt)])
    x.append(response.content)  

    return x
```

```python
columns =list(df.columns)
columns.append("Expanded Description")
new_df = pd.DataFrame(columns=columns)
for i in range(50):
    x = expand_description(df.iloc[i,:])
    new_df.loc[len(new_df)] = x
```
